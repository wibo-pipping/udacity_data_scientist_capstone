```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/wibo/Documents/education/udacity_data_scientist_capstone/test.csv', sep='\t')
df.head()
```

```python
import plotly.graph_objects as go
import plotly.express as px
```

```python
df.columns
```

```python
df.set_index('ds', inplace=True)
```

```python
df['type'] = 'historic'
df.loc[df.index[-180]:,'type'] = 'forecast'
```

```python
import datetime as dt
```

```python
dt.datetime.today().timestamp()*1000
```

```python
df.reset_index(inplace=True)
```

```python
dt.datetime.strptime(df.index[-180],'%Y-%m-%d').timestamp()*1000
```

```python
fig = px.line(df,
              x='ds',
              y='7d_rolling_avg',
              title=f'Stock price between {df.index.min()} and 2021-09-05, forecasted for 180 days',
              labels={'ds':'Date','7d_rolling_avg': 'Stock adjusted close ($)'},
              color_discrete_sequence=['#7d0c5b']
             )

fig.update_traces(hovertemplate = '<b>Historic data</b><br>Date: %{x}<br>Stock adjusted close: $%{y}',
                  name='Historic data',
                  showlegend=True
                 )

fig.add_vline(x=dt.datetime.today().timestamp()*1000,
              line={
                  'width':1.5,
                  'dash':'dot',
                  'color':'#e377c2'
              },
              annotation_text='hello'
             )

fig.add_trace(
    go.Scatter(
        x=df.iloc[-180:].index,
        y=df.iloc[-180:].yhat,
        mode='markers',
        marker={
            'color': '#ebb2da',
            'symbol': 'hexagon2',
            'size': 10
        },
        opacity=1,
        name='Forecasted value',
        hovertemplate = '<b>Forecasted</b><br>Date: %{x}<br>Stock adjusted close: $%{y}'
    )
)





fig.show()
```
